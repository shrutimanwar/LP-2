```python
import networkx as nx
import matplotlib.pyplot as plt

# Input number of vertices and edges
n = int(input("Enter number of vertices: "))
e = int(input("Enter number of edges: "))

# Create adjacency list
graph = {i: [] for i in range(n)}

print("Enter edges (e.g., 0 1):")
for _ in range(e):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)  # Undirected graph

# Visualize the graph using networkx
G = nx.Graph()
for node in graph:
    for neighbor in graph[node]:
        G.add_edge(node, neighbor)

plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightgreen', node_size=800, font_size=14)
plt.title("Undirected Graph - DFS")
plt.show()

# DFS recursive function
def dfs_recursive(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=' ')
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited)
    return visited

# Input for DFS start node
start_node = int(input("Enter starting node for DFS: "))
print("\nDFS Traversal (Recursive):")
dfs_recursive(graph, start_node)
```
